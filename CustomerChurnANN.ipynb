```python
df = df.dropna()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
for col in df:
  print(f'{col} : {df[col].unique()}')
```

```python
df.replace('No phone service', 'No', inplace=True)
df.replace('No internet service', 'No', inplace=True)
```

```python
for col in df:
  print(f'{col} : {df[col].unique()}')
```

```python
yn_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']

for col in yn_columns:
  df[col].replace({'Yes': 1, 'No': 0}, inplace=True)
```

```python
for col in df:
  print(f'{col} : {df[col].unique()}')
```

```python
df=pd.get_dummies(data=df, columns=['gender','InternetService','Contract','PaymentMethod'])
df.dtypes
```

```python
for col in df:
  print(f'{col} : {df[col].unique()}')
```

```python
cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```

```python
df[cols_to_scale].dtypes
```

```python
for col in df:
  print(f'{col} : {df[col].unique()}')
```

```python
X = df.drop(['Churn'], axis='columns')
y=df.Churn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
     keras.layers.Dense(10, input_shape=(27,), activation='relu'),
     keras.layers.Dense(1,activation='sigmoid')

 ])

model.compile(optimizer='adam',
               loss='binary_crossentropy',
               metrics=['accuracy'])

```

```python
y_columns = ['int']

for col in y_columns:
  df[col].replace({True: 1, False: 0}, inplace=True)
```

```python
X_train = X_train.astype('float32')
y_train = y_train.astype('float32')

model.fit(X_train, y_train, epochs=50)
```

```python
X_train.dtypes
```

```python
X_test = X_test.astype('float32')
y_test = y_test.astype('float32')
model.evaluate(X_test,y_test)
```
