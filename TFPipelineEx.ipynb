```python
def get_label(file_path):
  parts = tf.strings.split(file_path, os.path.sep)
  return parts[-2]
```

```python
def decode_text(file_path):
  label = get_label(file_path)
  return tf.io.read_file(file_path), label
```

```python
train_size = int(len(texts_ds) * 0.8)
train_ds = texts_ds.take(train_size)
test_ds= texts_ds.skip(train_size)
```

```python
train_ds = train_ds.map(decode_text)
```
